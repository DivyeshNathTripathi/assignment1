Q1. Explain with an example each when to use a for loop and a while loop.

ANS1:

 **For Loop**: Use a for loop when you know in advance how many times you want to iterate. For example, if you want to iterate over a list of items:


```python
fruits = ["apple", "banana", "cherry"]
for fruit in fruits:
    print(fruit)
```

**While Loop**: Use a while loop when the number of iterations is not known and depends on a condition. For example, if you want to keep prompting the user until they enter a specific value:

```python
number = 0
while number != 5:
    number = int(input("Enter a number (5 to stop): "))
```

**Q2**. Write a python program to print the sum and product of the first 10 natural numbers using for
and while loop.

Ans 2: 
Program to Print the Sum and Product of the First 10 Natural Numbers
Using a for loop:

```python
sum_of_numbers = 0
product_of_numbers = 1

for i in range(1, 11):
    sum_of_numbers += i
    product_of_numbers *= i

print("Sum:", sum_of_numbers)
print("Product:", product_of_numbers)
```

Using a while loop:

```python
sum_of_numbers = 0
product_of_numbers = 1
i = 1

while i <= 10:
    sum_of_numbers += i
    product_of_numbers *= i
    i += 1

print("Sum:", sum_of_numbers)
print("Product:", product_of_numbers)
```

**Q3**. Create a python program to compute the electricity bill for a household.

Ans 3: Program to Compute the Electricity Bill for a Household

```python
def calculate_bill(units):
    if units <= 100:
        bill = units * 4.5
    elif units <= 200:
        bill = (100 * 4.5) + (units - 100) * 6
    elif units <= 300:
        bill = (100 * 4.5) + (100 * 6) + (units - 200) * 10
    else:
        bill = (100 * 4.5) + (100 * 6) + (100 * 10) + (units - 300) * 20
    return bill


units = int(input("Enter the units of electricity consumed in a month: "))


total_bill = calculate_bill(units)


print("Total Electricity Bill: Rs.", total_bill)
```

**Q4.** Create a list of numbers from 1 to 100. Use for loop and while loop to calculate the cube of each
number and if the cube of that number is divisible by 4 or 5 then append that number in a list and print
that list.

Ans 4: 
Using a for loop:

```python

cubes_divisible = []

for number in range(1, 101):
    cube = number ** 3  
    if cube % 4 == 0 or cube % 5 == 0:  
        cubes_divisible.append(number) 

print("Numbers whose cubes are divisible by 4 or 5:", cubes_divisible)
```

Using a while loop:

```python

cubes_divisible = []

number = 1


while number <= 100:
    cube = number ** 3  
    if cube % 4 == 0 or cube % 5 == 0:  
        cubes_divisible.append(number)  
    number += 1  


print("Numbers whose cubes are divisible by 4 or 5:", cubes_divisible)
```

**Q5.** Write a program to filter count vowels in the below-given string.
string = "I want to become a data scientist"

Ans 5:  program to count the vowels in the given string:

```python
# Function to count vowels in a string
def count_vowels(s):
    vowels = 'aeiouAEIOU'  # Vowel characters (both lowercase and uppercase)
    count = 0  # Initialize vowel count to 0
    for char in s:
        if char in vowels:
            count += 1  # Increment count if the character is a vowel
    return count

# Given string
string = "I want to become a data scientist"

# Call the function and print the vowel count
vowel_count = count_vowels(string)
print("Number of vowels:", vowel_count)
```
